```python
print('\033[1;31;40m' + 'Hey, I am Random Forest Algorithm!' + '\033[0m')

import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_predict
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, auc
import matplotlib.pyplot as plt
import time

# Importing the dataset
dataset = pd.read_csv('thedataset.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# Feature Scaling
sc = StandardScaler()
X_scaled = sc.fit_transform(X)

# Training the Random Forest Classification model
classifier = RandomForestClassifier(n_estimators=20, criterion='entropy', random_state=0)

conf_matrices = []
accuracies = []
performance_metrics = []
roc_auc_scores = []

# Perform 10-fold cross-validation
for epoch in range(10):
    start_epoch_time = time.time()  # Start time for the current epoch

    # Perform cross-validation
    y_pred = cross_val_predict(classifier, X_scaled, y, cv=10)

    # Calculate confusion matrix and accuracy for the current epoch
    cm = confusion_matrix(y, y_pred)
    accuracy = accuracy_score(y, y_pred)

    # Calculate performance metrics for the current epoch
    tpr = cm[1][1] / (cm[1][1] + cm[1][0])
    tnr = cm[0][0] / (cm[0][0] + cm[0][1])
    fpr = cm[0][1] / (cm[0][1] + cm[0][0])
    fnr = cm[1][0] / (cm[1][0] + cm[1][1])
    precision = cm[1][1] / (cm[1][1] + cm[0][1])
    f1_score = (2 * tpr * precision) / (tpr + precision)
    balanced_accuracy = (tpr + tnr) / 2
    tss = tpr - fpr
    hss = (2 * ((cm[0][0] * cm[1][1]) - (cm[0][1] * cm[1][0]))) / ((cm[0][0] + cm[1][0]) * (cm[1][0] + cm[1][1]) + (cm[0][0] + cm[0][1]) * (cm[0][1] + cm[1][1]))
    brier_score = (((cm[0][1] + cm[1][0])**2 + (cm[1][1] + cm[0][0])**2) / np.sum(cm)**2)
    bss = 1 - (brier_score / 0.25)
    
    # Store confusion matrix, accuracy, and performance metrics for later averaging
    conf_matrices.append(cm)
    accuracies.append(accuracy)
    performance_metrics.append([tpr, tnr, fpr, fnr, precision, f1_score, balanced_accuracy, tss, hss, brier_score, bss, accuracy])

    # Display performance metrics for the current epoch
    print(f"\nEpoch {epoch + 1} Performance Metrics:")
    results = pd.DataFrame({
        "Metric": ["True Positive Rate (TPR)", "True Negative Rate (TNR)", "False Positive Rate (FPR)",
                   "False Negative Rate (FNR)", "Precision", "F1 Score", "Balanced Accuracy",
                   "True Skill Statistics (TSS)", "Heidke Skill Score (HSS)", "Brier Score",
                   "Brier Skill Score (BSS)", "Accuracy"],
        "Value": [tpr, tnr, fpr, fnr, precision, f1_score, balanced_accuracy, tss, hss, brier_score, bss, accuracy]
    })
    print(results)

    # Display execution time for the current epoch
    epoch_execution_time = time.time() - start_epoch_time
    print(f"\nExecution Time for Epoch {epoch + 1}: {epoch_execution_time:.2f} sec\n")

    # If it's the first epoch, calculate ROC AUC and plot ROC curve
    if epoch == 0:
        y_pred_prob = cross_val_predict(classifier, X_scaled, y, cv=10, method='predict_proba')
        fpr, tpr, thresholds = roc_curve(y, y_pred_prob[:, 1])
        roc_auc = auc(fpr, tpr)
        
        # Plot ROC curve for the first epoch
        plt.figure()
        lw = 2
        plt.plot(fpr, tpr, color='darkorange', lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
        plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (First Epoch)')
        plt.legend(loc="lower right")
        plt.show()
        
        # Display the ROC AUC score for the first epoch
        print("\nROC AUC for the First Epoch:", roc_auc)

# Calculate average confusion matrix, accuracy, and performance metrics
avg_cm = np.mean(conf_matrices, axis=0)
avg_accuracy_rf = np.mean(accuracies)
avg_performance_metrics_rf = np.mean(performance_metrics, axis=0)

# Display the average confusion matrix
print("Average Confusion Matrix:")
print(pd.DataFrame(avg_cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]))

# Display the average accuracy
print("\nAverage Accuracy:", avg_accuracy_rf)

# Display the average performance metrics
avg_results_rf = pd.DataFrame({
    "Metric": ["Average True Positive Rate (TPR)", "Average True Negative Rate (TNR)", "Average False Positive Rate (FPR)",
               "Average False Negative Rate (FNR)", "Average Precision", "Average F1 Score", "Average Balanced Accuracy",
               "Average True Skill Statistics (TSS)", "Average Heidke Skill Score (HSS)", "Average Brier Score",
               "Average Brier Skill Score (BSS)", "Average Accuracy"],
    "Value": avg_performance_metrics_rf
})
print("\nAverage Performance Metrics:")
print(avg_results_rf)


print('\033[1;31;40m' + 'Hey, I am KNN Algorithm!' + '\033[0m')

import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_predict, cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import time

# Importing the dataset
dataset = pd.read_csv('thedataset.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Training the K-NN model on the Training set
classifier = KNeighborsClassifier(n_neighbors=7, metric='euclidean', p=2)

conf_matrices = []
accuracies = []
performance_metrics = []

# Perform 10-fold cross-validation
for fold in range(10):
    start_fold_time = time.time()  # Start time for the current fold
    y_pred = cross_val_predict(classifier, X, y, cv=10)
    cm = confusion_matrix(y, y_pred)
    conf_matrices.append(cm)
    accuracy = accuracy_score(y, y_pred)
    accuracies.append(accuracy)
    fold_execution_time = time.time() - start_fold_time  # Calculate time taken for current fold

    # Calculating performance metrics for the current fold
    TPR = cm[1][1] / (cm[1][1] + cm[1][0])
    TNR = cm[0][0] / (cm[0][0] + cm[0][1])
    FPR = cm[0][1] / (cm[0][1] + cm[0][0])
    FNR = cm[1][0] / (cm[1][0] + cm[1][1])
    Precision = cm[1][1] / (cm[1][1] + cm[0][1])
    F1_Score = 2 * (Precision * TPR) / (Precision + TPR)
    Balanced_Accuracy = (TPR + TNR) / 2
    TSS = TPR - FPR
    HSS = (2 * ((cm[0][0] * cm[1][1]) - (cm[0][1] * cm[1][0]))) / ((cm[0][0] + cm[1][0]) * (cm[1][0] + cm[1][1]) + (cm[0][0] + cm[0][1]) * (cm[0][1] + cm[1][1]))
    BS = ((cm[0][1] + cm[1][0])**2 + (cm[1][1] + cm[0][0])**2) / np.sum(cm)**2
    BS_r = 0.25
    BSS = 1 - (BS / BS_r)
    
    # Store performance metrics for later averaging
    performance_metrics.append([TPR, TNR, FPR, FNR, Precision, F1_Score, Balanced_Accuracy, TSS, HSS, BS, BSS, accuracy])

    # Displaying performance metrics for the current fold
    print(f"\nFold {fold + 1} Performance Metrics:")
    results = pd.DataFrame({
        "Metric": ["True Positive Rate (TPR)", "True Negative Rate (TNR)", "False Positive Rate (FPR)",
                   "False Negative Rate (FNR)", "Precision", "F1 Score", "Balanced Accuracy",
                   "True Skill Statistics (TSS)", "Heidke Skill Score (HSS)", "Brier Score",
                   "Brier Skill Score (BSS)", "Accuracy"],
        "Value": [TPR, TNR, FPR, FNR, Precision, F1_Score, Balanced_Accuracy,
                  TSS, HSS, BS, BSS, accuracy]
    })
    print(results)

    # Calculate ROC curve and AUC score for just the first fold
    if fold == 0:
        y_probs = cross_val_predict(classifier, X, y, cv=10, method='predict_proba')
        fpr, tpr, thresholds = roc_curve(y, y_probs[:, 1])
        auc_score = roc_auc_score(y, y_probs[:, 1])

        # Plot ROC curve
        plt.figure()
        plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % auc_score)
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic Curve')
        plt.legend(loc="lower right")
        plt.show()

        # Print AUC score
        print("AUC Score:", auc_score)

    # Display execution time for the current fold
    print(f"\nExecution Time for Fold {fold + 1}: {fold_execution_time:.2f} sec\n")

# Calculate average confusion matrix, accuracy, and performance metrics
avg_cm = np.mean(conf_matrices, axis=0)
avg_accuracy_knn = np.mean(accuracies)
avg_performance_metrics_knn = np.mean(performance_metrics, axis=0)

avg_cm = avg_cm.astype(int)

# Display the average confusion matrix
print("Average Confusion Matrix:")
print(pd.DataFrame(avg_cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]))

# Display the average accuracy
print("\nAverage Accuracy:", avg_accuracy_knn)

# Display the average performance metrics
avg_results_knn = pd.DataFrame({
    "Metric": ["Average True Positive Rate (TPR)", "Average True Negative Rate (TNR)", "Average False Positive Rate (FPR)",
               "Average False Negative Rate (FNR)", "Average Precision", "Average F1 Score", "Average Balanced Accuracy",
               "Average True Skill Statistics (TSS)", "Average Heidke Skill Score (HSS)", "Average Brier Score",
               "Average Brier Skill Score (BSS)", "Average Accuracy"],
    "Value": avg_performance_metrics_knn
})
print("\nAverage Performance Metrics:")
print(avg_results_knn)

print('\033[1;31;40m' + 'Hey, I am LSTM Algorithm!' + '\033[0m')

import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, auc
import time
from tabulate import tabulate
import matplotlib.pyplot as plt

# Define a wrapper class around Keras model
class LSTMClassifierWrapper():
    def __init__(self):
        self.model = self.create_model()

    def create_model(self):
        model = tf.keras.Sequential([
            tf.keras.layers.LSTM(units=6, input_shape=(X_scaled.shape[1], 1)),
            tf.keras.layers.Dense(units=1, activation='sigmoid')
        ])
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        return model

    def fit(self, X, y, verbose=0):
        X_lstm = X.reshape(X.shape[0], X.shape[1], 1)  # Reshape for LSTM input
        history = self.model.fit(X_lstm, y, epochs=10, batch_size=32, verbose=verbose)
        return history

    def predict(self, X):
        if self.model is None:
            raise ValueError("Model is not trained. Call fit() before making predictions.")
        if len(X) == 0:
            raise ValueError("Input data is empty.")

        X_lstm = X.reshape(X.shape[0], X.shape[1], 1)  # Reshape for LSTM input
        predictions = (self.model.predict(X_lstm) > 0.5).astype(int)
        return predictions

# Load the dataset
dataset = pd.read_csv('thedataset.csv')  # Make sure to provide the correct path

# Splitting the dataset into features and target
X = dataset[['Age', 'EstimatedSalary']].values
y = dataset['Purchased'].values

# Feature Scaling
sc = StandardScaler()
X_scaled = sc.fit_transform(X)

# Initialize the model
wrapper_model = LSTMClassifierWrapper()

# Define number of folds
num_folds = 10
kf = KFold(n_splits=num_folds)

# Initialize a list to store results for each fold
fold_results = []

# Start cross-validation
fold_number = 1
for train_index, test_index in kf.split(X_scaled):
    fold_result = {}
    fold_result['Fold'] = fold_number
    print(f"Fold {fold_number}:")

    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Train the model
    start_time = time.time()
    history = wrapper_model.fit(X_train, y_train, verbose=1)
    end_time = time.time()
    training_time = end_time - start_time

    # Predictions
    y_pred = wrapper_model.predict(X_test)

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()

    # Performance Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1_score = 2 * (precision * recall) / (precision + recall)
    balanced_accuracy = 0.5 * ((tp / (tp + fn)) + (tn / (tn + fp)))
    true_skill_score = (tp / (tp + fn)) - (fp / (fp + tn))
    heidke_skill_score = (2 * ((tp * tn) - (fp * fn))) / (((tp + fn) * (fn + tn)) + ((tp + fp) * (fp + tn)))
    brier_score = ((tp + tn) / (tp + tn + fp + fn)) ** 2 + ((fp + fn) / (tp + tn + fp + fn)) ** 2
    brier_skill_score = 1 - (brier_score / ((fp + fn) / (tp + tn)))

    # Append metrics to the fold result
    fold_result['Epochs'] = len(history.history['accuracy'])
    fold_result['Time (s)'] = training_time
    fold_result['Accuracy'] = accuracy
    fold_result['TPR'] = tp / (tp + fn)
    fold_result['TNR'] = tn / (tn + fp)
    fold_result['FPR'] = fp / (fp + tn)
    fold_result['FNR'] = fn / (fn + tp)
    fold_result['Precision'] = precision
    fold_result['F1 Score'] = f1_score
    fold_result['Balanced Accuracy'] = balanced_accuracy
    fold_result['True Skill Score'] = true_skill_score
    fold_result['Heidke Skill Score'] = heidke_skill_score
    fold_result['Brier Score'] = brier_score
    fold_result['Brier Skill Score'] = brier_skill_score

    # Print results
    print(tabulate([fold_result], headers='keys', tablefmt='pretty'))

    # Calculate ROC curve and AUC for the first epoch only
    if fold_number == 1:
        y_pred_prob = wrapper_model.model.predict(X_test.reshape(X_test.shape[0], X_test.shape[1], 1))
        fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
        roc_auc = auc(fpr, tpr)

        # Plot ROC curve
        plt.figure()
        plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (First Epoch)')
        plt.legend(loc="lower right")
        plt.show()

    # Add fold result to fold_results list
    fold_results.append(fold_result)

    # Increment fold number
    fold_number += 1

# Calculate average performance metrics
average_results_lstm = {}
for key in fold_results[0].keys():
    if key != 'Fold':
        average_results_lstm[key] = np.mean([fold[key] for fold in fold_results])

# Print average performance metrics
print("\nAverage Performance Metrics Across Folds:")
print(tabulate([average_results_lstm], headers='keys', tablefmt='pretty'))


from tabulate import tabulate

# Create a dictionary to store the performance metrics of each algorithm
print('\033[1;31;40m' + 'Comparing The 3 Algorithms' + '\033[0m')
performance_metrics = {
    "Algorithm": ["Random Forest", "KNN", "LSTM"],
    "Average Accuracy": [avg_accuracy_rf, avg_accuracy_knn, average_results_lstm['Accuracy']],
    "Average True Positive Rate (TPR)": [avg_performance_metrics_rf[0], avg_performance_metrics_knn[0], average_results_lstm['TPR']],
    "Average True Negative Rate (TNR)": [avg_performance_metrics_rf[1], avg_performance_metrics_knn[1], average_results_lstm['TNR']],
    "Average False Positive Rate (FPR)": [avg_performance_metrics_rf[2], avg_performance_metrics_knn[2], average_results_lstm['FPR']],
    "Average False Negative Rate (FNR)": [avg_performance_metrics_rf[3], avg_performance_metrics_knn[3], average_results_lstm['FNR']],
    "Average Precision": [avg_performance_metrics_rf[4], avg_performance_metrics_knn[4], average_results_lstm['Precision']],
    "Average F1 Score": [avg_performance_metrics_rf[5], avg_performance_metrics_knn[5], average_results_lstm['F1 Score']],
    "Average Balanced Accuracy": [avg_performance_metrics_rf[6], avg_performance_metrics_knn[6], average_results_lstm['Balanced Accuracy']],
    "Average True Skill Statistics (TSS)": [avg_performance_metrics_rf[7], avg_performance_metrics_knn[7], average_results_lstm['True Skill Score']],
    "Average Heidke Skill Score (HSS)": [avg_performance_metrics_rf[8], avg_performance_metrics_knn[8], average_results_lstm['Heidke Skill Score']],
    "Average Brier Score": [avg_performance_metrics_rf[9], avg_performance_metrics_knn[9], average_results_lstm['Brier Score']],
    "Average Brier Skill Score (BSS)": [avg_performance_metrics_rf[10], avg_performance_metrics_knn[10], average_results_lstm['Brier Skill Score']]
}

# Convert the dictionary to a DataFrame
performance_df = pd.DataFrame(performance_metrics)

# Transpose the DataFrame
performance_df = performance_df.T

# Print the transposed DataFrame
performance_df.columns = performance_df.iloc[0]
performance_df = performance_df.drop(performance_df.index[0])
performance_df
```
